```python
# output pic

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# read input csv ==================
direct_ratio_SL_aucc_list = pd.read_csv("../figure/sigir/SL.csv")
direct_ratio_XL_aucc_list = pd.read_csv("../figure/sigir/TPM-XL.csv")

grf = pd.read_csv("../figure/sigir/CF.csv")

dr = pd.read_csv("../figure/sigir/DRM.csv")


DRP_avg_aucc_cost_curve = pd.read_csv("../figure/sigir/DRP.csv")



rDRP_avg_aucc_cost_curve = pd.read_csv("../figure/sigir/DCL-new.csv")






x_index = DRP_avg_aucc_cost_curve['delta_cost'].values
y_random = DRP_avg_aucc_cost_curve['delta_reward'].values[-1] / x_index[-1] * x_index

x_normalization = x_index[-1]
y_normalization = DRP_avg_aucc_cost_curve['delta_reward'].values[-1]


# random ===================================
plt.plot(x_index / x_normalization, y_random / y_normalization, color='#000000', label='Random', linewidth=1.5)

# TPM-SL
plt.plot(x_index / x_normalization, direct_ratio_SL_aucc_list['delta_reward'].values / y_normalization, color='#1f77b4', label='TPM-SL', linewidth=1.5)


# TPM-XL
plt.plot(x_index / x_normalization, direct_ratio_XL_aucc_list['delta_reward'].values / y_normalization, color='#808000', label='TPM-XL', linewidth=1.5)


# TPM-CF
plt.plot(x_index / x_normalization, grf['delta_reward'].values / y_normalization, color='#800080', label='TPM-CF', linewidth=1.5)


# DR
plt.plot(x_index / x_normalization, dr['delta_reward'].values / y_normalization, color='#d62728', label='DRM', linewidth=1.5)



# DRP
plt.plot(x_index / x_normalization, DRP_avg_aucc_cost_curve['delta_reward'].values / y_normalization, color='#FBB454', label='DRP', linewidth=1.5)




# rDPR
plt.plot(x_index / x_normalization, rDRP_avg_aucc_cost_curve['delta_reward'].values / y_normalization, color='#40a368', label='DCL', linewidth=1.5)


#plt.title('Sufficient data and No covariate shift')


plt.xlabel('Incremental cost', fontsize=12, fontweight='bold')
plt.ylabel('Incremental reward', fontsize=12, fontweight='bold')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend()
plt.grid(True)
plt.tight_layout()


# Save and display the figure

plt.savefig('../figure/sigir/offline_pic.pdf', format='pdf', bbox_inches = 'tight')
plt.show()
```
